# ■ 다차원 배열계산

다차원 배열도 그 기본은 '숫자의 집합' 입니다.  
숫자를 한 줄로 늘어선 것이나 정사각형으로 늘어놓은 것이나
3차원으로 늘어놓은 것이나 전부 다차원 배열이라고 합니다.

## 예제1. 1차원 배열을 만드는 방법

```python
import numpy as np
a = np.array([1, 2, 3, 4])
print( np.ndim(a) )  # 차원을 확인
```

위의 내적을 총 3가지 방법으로 구현하시오

## 1. 첫번째 방법

```python
import numpy as np

x = np.array([ [1,2], [3,4] ])
y = np.array([ [5,6], [7,8] ])
print( np.dot(x,y) )
```

## 2. 두번째 방법

```python
import numpy as np

x = np.matrix([ [1,2], [3,4] ])
y = np.matrix([ [5,6], [7,8] ])
print( x*y )
```

## 3. 세번째 방법

```python
import numpy as np

x = np.array([ [1,2], [3,4] ])
y = np.array([ [5,6], [7,8] ])
print( x @ y )
```

## ※ array 와 matrix 의 기능 차이 ?

array 는 다차원으로 나타낼 수 있는데 matrix 는 2차원 밖에 안된다.

# ■ 3층 신경망 구현하기

<div>

</div>

## 1. 입력층 -----> 은닉1층

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x) )

x = np.array([1,2])
w1 = np.array([[1,3,5], [2,4,6]])
y = np.dot(x,w1)
y_hat = sigmoid(y)
print(y_hat)
```

## 2. 입력층 ----> 은닉1층 ----> 은닉2층

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x) )

# 0층
x = np.array( [1,2] )

# 1층
w1 = np.array([[1,3,5], [2,4,6]])
y = np.dot(x,w1)
y_hat = sigmoid(y)

# 2층
w2 = np.array([[3,4], [5,6], [7,8]])
z = np.dot(y_hat, w2)
z_hat = sigmoid(z)
print(z_hat)
```

## 문제45. 아래의 신경망의 k 값 까지 출력하세요

<div>

</div>

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x) )

# 0층
x = np.array( [1,2] )

# 1층
w1 = np.array([[1,3,5], [2,4,6]])
y = np.dot(x,w1)
y_hat = sigmoid(y)

# 2층
w2 = np.array([[3,4], [5,6], [7,8]])
z = np.dot(y_hat, w2)
z_hat = sigmoid(z)

# 3층
w3 = np.array([[4,5], [6,7]])
k = np.dot(z_hat, w3)
print(k)
```
