위의 내적을 총 3가지 방법으로 구현하시오

## 1. 첫번째 방법

```python
import numpy as np

x = np.array([ [1,2], [3,4] ])
y = np.array([ [5,6], [7,8] ])
print( np.dot(x,y) )
```

## 2. 두번째 방법

```python
import numpy as np

x = np.matrix([ [1,2], [3,4] ])
y = np.matrix([ [5,6], [7,8] ])
print( x*y )
```

## 3. 세번째 방법

```python
import numpy as np

x = np.array([ [1,2], [3,4] ])
y = np.array([ [5,6], [7,8] ])
print( x @ y )
```

## ※ array 와 matrix 의 기능 차이 ?

array 는 다차원으로 나타낼 수 있는데 matrix 는 2차원 밖에 안된다.

## ※ 행렬곱(내적) 시 주의 사항 !

<div>

</div>

다차원 배열을 곱하려면 두 행렬의 대응하는 차원의 원소 수를 일치시켜야 합니다.
