			Rectified : 정류된

정류는 전기회로 용어로, 예를 들어 반파정류회로는 +/- 가 반복되는 교류에서
 - 흐름을 차단하는 회로입니다.

Relu 함수 그래프(그림3-9 p76) 처럼 x 가 0 이하 일때를 차단하여 아무런 값도
출력하지 않는 (0을 출력) 것 입니다.

# 예제1. Relu 함수를 생성하시오 !

" Relu 함수는 0 이하는 0으로 흘려보내고 입력값이 0 이 넘으면 입력값 그대로 흘려보내는 함수 "

```python
import numpy as np

def relu(x):
    return np.maximum(0, x)  # 0 과 x 값 중에 큰 값을 출력하시오 !

print( relu(-2) )
print( relu(0.3) )
```

# 문제39. 위의 relu 함수를 그래프로 시각화 하시오 !

```python
import matplotlib.pyplot as plt
import numpy as np

def relu(x):
    return np.maximum(0, x)

x = np.arange(-5, 5, 0.1)  
y = relu(x)

plt.plot(x,y)
plt.show()
```
