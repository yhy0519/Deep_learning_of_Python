# ■ 다차원 배열계산

다차원 배열도 그 기본은 '숫자의 집합' 입니다.
숫자를 한 줄로 늘어선 것이나 정사각형으로 늘어놓은 것이나
3차원으로 늘어놓은 것이나 전부 다차원 배열이라고 합니다.

## 예제1. 1차원 배열을 만드는 방법

```python
import numpy as np
a = np.array([1, 2, 3, 4])
print( np.ndim(a) )  # 차원을 확인
```

# 문제41. 3차원 배열을 만들어 보시오 !

```python
import numpy as np

b = np.array([ [ [1,2], [3,4] ], [ [5,6], [7,8] ] ])
print(b)
print(np.ndim(b))
```

# 문제42. 아래의 2차원 배열에서 숫자 3을 출력하시오 !

$$\begin{pmatrix} 1 & 2 \\ 3 & 4 \\ 5 & 6 \end{pmatrix}$$

```python
import numpy as np

b = np.array([ [1,2], [3,4], [5,6] ])
print(b)
print(np.ndim(b)) 
print(b[1][0])
```

# 문제43. 아래의 3차원 배열에서 숫자 5를 출력하시오 !

```python
import numpy as np

b = np.array([ [ [1,2], [3,4] ], [ [5,6], [7,8] ] ])
print(b[1][0][0])
```
